```python
import matplotlib.pyplot as plt #파이썬으로 데이터를 시각화 하는데 matploylib 라이브러리를 가장 많이씀. 
#pyplot 모듈은 공학용 도구로 알려진 MATLAB과 사용법이 유사하다.
plt.title('plotting') #제목

plt.plot([10,20,30,40],color='skyblue',linestyle='--',label='asc') #plot() 함수는 직선 또는 꺾은선 형태의 그래프를 그릴때 사용할 수 있는 명령어임 #증가의 의미 asc 범례
plt.plot([50,40,20,10],'r.',label='asc') # r. 점모양 마커 , g^ 삼각형 
plt.plot([40,30,20,10],color='pink',linestyle='--',label='desc')# 감소의 의미 desc 범례 # 그래프 선양 바꾸기 linestyle
plt.legend() #범례 표시
plt.show()
```

```python
# 특정 날짜 기온 변화를 그래프로 그림
import csv
import matplotlib.pyplot as plt

f=open('C:/Users/kimhk/Downloads/DataAnalysis/jupyter/CsvRead-DataAnalysis/seoul.csv')
data = csv.reader(f)
next(data)
high=[] #최고 기온값을 저장할 리스트 high 
low=[] #최저 기온값을 저장할 리스트 low

for row in data :
    if row[-1] !='' and row[-2] !='': #최고 기온값과 최저 기온 값이 존재하면
        date = row[0].split('-') #날짜값을 -문자를 기준으로 구분하여 저장
        if 1991 <= int(date[0]) : #1983년 이후 데이터라면
            if date[1] == '10' and date[2] == '30': #10월 30일
                high.append(float(row[-1])) # 최고 기온값을 high 리스트 저장
                low.append(float(row[-2])) # 최저 기온값을 low 리스트 저장
plt.rc('font',family="Malgun Gothic") #맑은 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] =False # 마이너스 기호 깨짐방지
plt.title('10월 30일 기온 변화 그래프') #제목
#high,low 리스트에 저장된값 원하는 색, 레이블 표사
plt.plot(high,'red',label="최고 기온")
plt.plot(low,'skyblue',label="최저 기온")

plt.legend() #범례 표시
plt.show() #그래프 나타내기

                
```
