```python
import csv
import matplotlib.pyplot as plt


f = open('age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ') #입력 예시 : 신림
for row in data : 
    if name in row[0] :
        for i in row[3:] :
            result.append(int(i.replace(',',''))) # csv 안에 ,를 공백으로 처리해야 int 형으로 읽을수 있음
            
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title(name+'지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
#bar chart
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)

result = [] 

for row in data : 
    if '신도림' in row[0] :
        for i in row[3:] :
            result.append(int(i.replace(',',''))) # csv 안에 ,를 공백으로 처리해야 int 형으로 읽을수 있음
            
plt.bar(range(len(result)),result) # len(result) 배열의 길이를 구함
plt.show()
            
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')

data = csv.reader(f)

#g1 남자 g2 여자
g1=[]
g2=[]

name = input('찾고 싶은 지역의 이름을 입력해주세요:')

for row in data : 
    if name in row[0] : 
        for i in row[3:104] : 
            g1.append(-int(i.replace(',','')))
        for i in row[106:]:
            g2.append(int(i.replace(',','')))

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+'지역의 남녀 성별 인구 분포')
plt.barh(range(len(g1)),g1,label='남성')
plt.barh(range(len(g2)),g2,label='여성')
plt.legend()
plt.show()
```
